### 复数

```python
a = 1+1j
output = np.angle(a, deg=False)
output
```

 2. [ ] 返回`a`的实数和虚数部分值。

```python
a = np.array([1+2j, 3+4j, 5+6j])
real = np.real(a) # a.real
imag = np.imag(a) # a.imag
print("real part=", real)
print("imaginary part=", imag)
```

### 离散傅立叶变换 (DFT)

 1. [ ] 计算`a`的一个空间的DFT值。

```python
a = np.exp(2j * np.pi * np.arange(8))
output = np.fft.fft(a)
output
```

 2. [ ] 在上一题中`output`的逆序DFT值。

```python
print("a=", a)
inversed = np.fft.ifft(output)
assert np.allclose(np.fft.ifft(np.fft.fft(a)), a)
print("\ninversed=", a)
```

 3. [ ] 计算`a`实数输入值的一个空间DFT值。

```python
a = [0, 1, 0, 0]
output = np.fft.rfft(a)
print(output)
assert output.size == len(a) // 2 + 1 if len(a) % 2 == 0 else (len(a) + 1) // 2

# cf.
output2 = np.fft.fft(a)
print(output2)
```

 4. [ ] 计算上一道题中`output`的逆序DFT值。

```python
inversed = np.fft.ifft(output)
assert np.allclose(np.fft.fft(np.fft.ifft(a)), a)
print("inversed=", a)
```

 5. [ ] 返回`a`的DFT取样频率值。

```python
signal = np.array([-2, 8, 6, 4, 1, 0, 3, 5], dtype=np.float32)
fourier = np.fft.fft(signal)
n = signal.size
freq = np.fft.fftfreq(n, d=1)
freq
```

### 一些视窗函数

```python
from matplotlib.font_manager import FontProperties

font = FontProperties(fname=r'/Library/Fonts/Hanzipen.ttc')

fig = plt.figure(figsize=(19, 10))

# Hamming window
window = np.hamming(51)
plt.plot(np.bartlett(51), label="Bartlett window")
plt.plot(np.blackman(51), label="Blackman window")
plt.plot(np.hamming(51), label="Hamming window")
plt.plot(np.hanning(51), label="汉宁视窗")
plt.plot(np.kaiser(51, 14), label="凯泽视窗")
plt.xlabel("取样", fontproperties=font)
plt.ylabel("振幅", fontproperties=font)
plt.legend(prop=font)
plt.grid()

plt.show()
```
